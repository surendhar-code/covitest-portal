## **REMOVE UNNECESSARY DATA **


REMOVE 'OTHER' CATEGORY IN CORONA RESULT COLUMN.

## **REMOVE NONE CATEGORY IN A COLUMN**

```python
# create a function to remove none category in a column
def remove_none(col):
  indexes = covid_df[covid_df[col]=='None'].index
  return indexes
```

```python
cough_none = remove_none('cough')
covid_df.drop(cough_none,inplace=True)
```

```python
fever_none = remove_none('fever')
covid_df.drop(fever_none,inplace=True)
```

```python
sorethroat_none = remove_none('sore_throat')
covid_df.drop(sorethroat_none,inplace=True)
```

```python
breathing_none = remove_none('shortness_of_breath')
covid_df.drop(breathing_none,inplace=True)
```

```python
headache_none = remove_none('head_ache')
covid_df.drop(headache_none, inplace=True)
```

```python
for col in covid_df.columns:
  print(covid_df[col].value_counts())
  print('-------------------------------------------------------------------------------------------------------')
```

## **REPLACE 0 AND 1 IN A COLUMN.**
Since there are 2 zeros and 2 ones totally 4 categories in a dataset. We want only one 0 and one 1.

```python
# create a function to replace two zeros and two ones with one 0 and one 1.
def replace_zero_one(col):
 covid_df[col].replace(to_replace=0,value=0, inplace=True)
 covid_df[col].replace(to_replace=1,value=1, inplace=True)
 covid_df[col].replace(to_replace='0',value=0, inplace=True)
 covid_df[col].replace(to_replace='1',value=1, inplace=True)
```

```python
replace_zero_one('cough')
covid_df.cough.value_counts()
```

```python
replace_zero_one('fever')
covid_df.fever.value_counts()
```

```python
replace_zero_one('sore_throat')
covid_df.sore_throat.value_counts()
```

```python
replace_zero_one('shortness_of_breath')
covid_df.shortness_of_breath.value_counts()
```

```python
replace_zero_one('head_ache')
covid_df.head_ache.value_counts()
```

```python
# replace None with Other in Gender.
covid_df.gender.replace(to_replace='None',value='Other',inplace=True)
covid_df.gender.value_counts()
```

```python
# replace None with No in age above 60.
covid_df.age_60_and_above.replace(to_replace='None',value='No',inplace=True)
covid_df.age_60_and_above.value_counts()
```

#**ONE HOT ENCODING**

Applying one hot encoding to corona result column.


```python
# by using pandas get_dummies we can perform one hot encoding for corona result column
result_dummies = pd.get_dummies(covid_df['corona_result'])
```

```python
result_dummies.drop(columns=['negative'],inplace=True)
```

```python
result_dummies.columns = ['Result']
```

```python
result_dummies
```

```python
covid_df = pd.concat([covid_df,result_dummies],axis=1)
```

```python
covid_df.drop(columns=['corona_result'],inplace=True)
```

Apply one hot encoding to gender column.

```python
# by using pandas get_dummies we can perform one hot encoding for gender column
gender_dummies = pd.get_dummies(covid_df['gender'])
```

```python
gender_dummies.drop(columns=['Other'],inplace=True)
gender_dummies.columns = ['Female','Male']
```

```python
gender_dummies
```

```python
covid_df = pd.concat([covid_df,gender_dummies],axis=1)
```

```python
covid_df.drop(columns=['gender'],inplace=True)
```

```python
covid_df
```

Apply One Hot Encoding to age above 60 column.

```python
# by using pandas get_dummies we can perform one hot encoding for gender column
age60_dummies = pd.get_dummies(covid_df['age_60_and_above'])
age60_dummies
```

```python
age60_dummies.drop(columns=['No'],inplace=True)
age60_dummies
```

```python
covid_df = pd.concat([covid_df,age60_dummies],axis=1)
```

```python
covid_df.drop(columns=['age_60_and_above'],inplace=True)
```

```python
covid_df.rename(columns={"Yes": "Age above 60"},inplace=True)
```

Apply One Hot Encoding to test indication column.

```python
# by using pandas get_dummies we can perform one hot encoding for gender column
test_dummies = pd.get_dummies(covid_df['test_indication'])
```

```python
test_dummies.drop(columns=['Other'],inplace=True)
```

```python
test_dummies
```

```python
covid_df = pd.concat([covid_df,test_dummies],axis=1)
covid_df
```

```python
covid_df.drop(columns=['test_indication','test_date'],inplace=True)
```
